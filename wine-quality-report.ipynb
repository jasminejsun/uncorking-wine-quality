## Dataset, problem, and performance metrics

### Dataset chosen

For this project, we will be looking at the UCI Wine Quality dataset (https://archive.ics.uci.edu/dataset/186/wine+quality). The dataset contains red and white Vinho Verde wine samples and their respective physicochemical traits.

Features include:

- Fixed acidity
- Volatile acidity
- Citric acid
- Residual sugar
- Chlorides
- Free sulfur dioxide
- Total sulfur dioxide
- Density
- pH
- Sulphates
- Alcohol

The target variable is a quality variable, which is a score between 0 and 10.

### Why the dataset is challenging

First, wines with different quality scores can share very similar chemical profiles, so perfect separation is difficult/not possible. As well, many features are strongly correlated (e.g., density and alcohol), which motivates dimensionality reduction.

### Problem definition

Which chemical properties best predict wine quality (density, acidity, alcohol)? We want to predict whether a wine is “high quality” (label 1) or “lower quality” (label 0) from its chemical properties.

### Performance metric

Our primary performance metric is classification accuracy on the test set. We also report F1-score as a secondary metric to handle moderate class imbalance between high- and low-quality wines.
